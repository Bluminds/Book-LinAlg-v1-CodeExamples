```python
# Ejercicio 9
import numpy as np

def gaussian_elimination(A, b):
    n = len(b)
    for i in range(n):
        # Pivoteo parcial
        max_row = np.argmax(np.abs(A[i:n, i])) + i
        A[[i, max_row]] = A[[max_row, i]]
        b[[i, max_row]] = b[[max_row, i]]

        # Eliminación
        for j in range(i + 1, n):
            factor = A[j, i] / A[i, i]
            A[j, i:] = A[j, i:] - factor * A[i, i:]
            b[j] = b[j] - factor * b[i]

    # Sustitución regresiva
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = (b[i] - np.dot(A[i, i+1:], x[i+1:])) / A[i, i]
    return x


# Definimos la matriz de coeficientes y el vector de términos constantes
A = np.array([[2, 3], [1, 3]], dtype=float)
b = np.array([1, 4], dtype=float)

# Resolvemos el sistema de ecuaciones Ax = b
x = gaussian_elimination(A, b)

print("Solución usando Eliminación Gaussiana:")
print(x)
```

```python
# Ejercicio 10
import numpy as np

def jacobi_method(A, b, x0, tol=1e-10, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i, j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i, i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new
        x = x_new
    return x

# Definimos la matriz de coeficientes y el vector de términos constantes
A = np.array([[4, 2], [1, 4]], dtype=float)
b = np.array([11, 10], dtype=float)

# Estimación inicial
x0 = np.zeros(len(b))

# Resolvemos el sistema de ecuaciones Ax = b
x = jacobi_method(A, b, x0)

print("Solución usando Método de Jacobi:")
print(x)
```

```python
# Ejercicio 20
import numpy as np

# Definimos una matriz cuadrada
A = np.array([[4, -2], [2, 7]])

# Calculamos los autovalores y autovectores
eigenvalues, eigenvectors = np.linalg.eig(A)

# Verificamos la propiedad Av = (lambda)v para el primer autovector y autovalor
v = eigenvectors[:, 0]
lambda_var = eigenvalues[0]
Av = np.dot(A, v)
lambda_v = lambda_var * v

print("A * v:")
print(Av)
print("lambda * v:")
print(lambda_v)
print("Son iguales A * v y lambda * v?:", np.allclose(Av, lambda_v))
```

```python
# Ejercicio 22
import numpy as np

# Datos de ejemplo: cada fila es una muestra y cada columna es una característica
X = np.array([[5, 4], [5, 7], [2.2, 2.9], [1.9, 2.2], [3.2, 3.9], [2.3, 2.7], [2.0, 1.6], [2.0, 2.1], [1.5, 1.6], [1.1, 0.9]])

# Restamos la media de cada columna (característica)
X_meaned = X - np.mean(X, axis=0)

# Calculamos la matriz de covarianza
cov_matrix = np.cov(X_meaned, rowvar=False)

# Calculamos los autovalores y autovectores de la matriz de covarianza
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Ordenamos los autovectores por los autovalores más grandes
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Seleccionamos los k autovectores (componentes principales)
k = 1  # Seleccionamos 1 componente principal para este ejemplo
principal_components = eigenvectors[:, :k]

# Transformamos los datos
X_reduced = np.dot(X_meaned, principal_components)

print("Matriz de covarianza:")
print(cov_matrix)
print("Autovalores:")
print(eigenvalues)
print("Autovectores:")
print(eigenvectors)
print("Componentes principales seleccionados:")
print(principal_components)
print("Datos reducidos:")
print(X_reduced)
```

```python
# Ejercicio 25
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# Cargamos un conjunto de datos de imágenes de dígitos
digits = load_digits()
X = digits.data
y = digits.target

# Aplicamos PCA para reducir la dimensionalidad a 2 componentes
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X)

# Visualizamos los dígitos en el nuevo espacio reducido
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar()
plt.title('Dígitos en el espacio de componentes principales')
plt.show()
```

```python
# Ejercicio 26
import numpy as np

# Definimos una matriz de estado A para un sistema dinámico
A = np.array([[1.1, 0.1],
              [0.3, 0.6]])

# Definimos el vector de estado inicial
x = np.array([1, 0])

# Simulamos la evolución del sistema durante 10 pasos de tiempo
states = [x]
for _ in range(10):
    x = np.dot(A, x)
    states.append(x)

# Convertimos la lista de estados a un array NumPy para su análisis
states = np.array(states)
print("Evolución del sistema:")
print(states)
```

```python
# Ejercicio 29
import matplotlib.pyplot as plt
import numpy as np

# Definimos dos líneas de ecuaciones
x = np.linspace(-10, 10, 200)
y1 = 4 - x
y2 = 2 * x - 1

# Creamos el gráfico de las dos líneas
plt.plot(x, y1, label='x + y = 4')
plt.plot(x, y2, label='2x - y = 1')

# Encontramos el punto de intersección
A = np.array([[1, 1], [2, -1]])
b = np.array([4, 1])
solution = np.linalg.solve(A, b)

# Añadimos el punto de intersección al gráfico
plt.plot(solution[0], solution[1], 'ro', label='Solución')

# Configuramos el gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Solución de un Sistema de Ecuaciones Lineales')
plt.legend()
plt.grid()
plt.show()

# Imprimimos la solución encontrada
print("La solución del sistema de ecuaciones es:", solution)
```

```python
# Ejercicio 30
import matplotlib.pyplot as plt
import numpy as np

# Definimos los puntos de una cuadrícula
x = np.linspace(-1, 2, 10)
y = np.linspace(-1, 2, 10)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# Definimos una matriz de transformación (rotación + escalado)
theta = np.pi / 4  # 45 grados
scale = 1.5
transformation_matrix = scale * np.array([[np.cos(theta), -np.sin(theta)],
                                          [np.sin(theta), np.cos(theta)]])
transformed_points = transformation_matrix @ points

# Creamos el gráfico antes y después de la transformación
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.quiver(points[0, :], points[1, :], angles='xy', scale_units='xy', scale=1)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title('Puntos Originales')
plt.grid()
plt.subplot(1, 2, 2)
plt.quiver(transformed_points[0, :], transformed_points[1, :], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title('Puntos Transformados')
plt.grid()
plt.show()
```
