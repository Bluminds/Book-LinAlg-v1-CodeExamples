```python
# Ejercicio 11
import numpy as np
import matplotlib.pyplot as plt

# Definir el vector
v = np.array([3, 3])

# Calcular la dirección del vector (ángulo con respecto al eje x)
theta = np.arctan2(v[1], v[0])  # en radianes
theta_deg = np.degrees(theta)  # convertir a grados
print(f"La dirección del vector {v} es {theta_deg:.2f} grados")

# Graficar el vector y el ángulo
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v')
plt.text(v[0]/2, v[1]/2, f'{theta_deg:.2f} grados', fontsize=12, color='r')
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dirección de un Vector')
plt.legend()
plt.grid()
plt.show()
```

```python
# Ejercicio 12
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir el vector
v = np.array([2, 3, 4])

# Calcular la magnitud del vector
magnitude = np.linalg.norm(v)

# Calcular los ángulos de dirección en 3D
theta_x = np.arccos(v[0] / magnitude)
theta_y = np.arccos(v[1] / magnitude)
theta_z = np.arccos(v[2] / magnitude)
theta_x_deg = np.degrees(theta_x)
theta_y_deg = np.degrees(theta_y)
theta_z_deg = np.degrees(theta_z)
print(f"Los ángulos de dirección del vector {v} son: \n_theta_x: {theta_x_deg:.2f} grados \n_theta_y: {theta_y_deg:.2f} grados \n_theta_z: {theta_z_deg:.2f} grados")

# Graficar el vector en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='r', label='v')
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Sentido y Dirección de un Vector en 3D')
plt.legend()
plt.show()
```

```python
# Ejercicio 15
import numpy as np
import matplotlib.pyplot as plt

# Definir los vectores
u = np.array([2, 4])
v = np.array([4, 2])

# Calcular el producto punto
dot_product = np.dot(u, v)

# Calcular las magnitudes
magnitude_u = np.linalg.norm(u)
magnitude_v = np.linalg.norm(v)

# Calcular el ángulo entre los vectores
cos_theta = dot_product / (magnitude_u * magnitude_v)
theta = np.arccos(cos_theta)
print(f"El ángulo entre {u} y {v} es {np.degrees(theta):.2f} grados")

# Graficar los vectores
plt.figure()
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Interpretación Geométrica del Producto Punto')
plt.legend()
plt.grid()
plt.show()
```

```python
# Ejercicio 19
# Definición de los vectores u y v
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])

# Cálculo del producto cruz
cross_product = np.cross(u, v)

# Cálculo de la magnitud del producto cruz
magnitude = np.linalg.norm(cross_product)

# Visualización de los vectores y el paralelogramo
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Vectores u y v
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='v')

# Producto cruz
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='g', label='u x v')

# Dibujar el paralelogramo
vertices = np.array([[0, 0, 0], u, u+v, v])
ax.plot_trisurf(vertices[:, 0], vertices[:, 1], vertices[:, 2], color='yellow', alpha=0.5)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

print("Producto Cruz (u x v):", cross_product)
print("Magnitud del Producto Cruz:", magnitude)
```

```python
# Ejercicio 20
# Definición de los vectores u, v y w
u = np.array([1, 2, 3])
v = np.array([1, 5, 6])
w = np.array([2, 8, 9])

# Cálculo de los productos cruz
cross_product_uv_plus_w = np.cross(u, v + w)
cross_product_uv = np.cross(u, v)
cross_product_uw = np.cross(u, w)

# Verificación de la propiedad distributiva
distributed_cross_product = cross_product_uv + cross_product_uw

# Visualización de los vectores y productos cruz
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Vectores u, v y w
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='v')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='c', label='w')

# Producto cruz y sus componentes distribuidas
ax.quiver(0, 0, 0, cross_product_uv_plus_w[0], cross_product_uv_plus_w[1], cross_product_uv_plus_w[2], color='g', label='u x (v + w)')
ax.quiver(0, 0, 0, distributed_cross_product[0], distributed_cross_product[1], distributed_cross_product[2], color='y', label='(u x v) + (u x w)')
ax.set_xlim([-15, 15])
ax.set_ylim([-15, 15])
ax.set_zlim([-15, 15])
ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

print("Producto Cruz u x (v + w):", cross_product_uv_plus_w)
print("Distribución (u x v) + (u x w):", distributed_cross_product)
```

```python
# Ejercicio 25
import numpy as np

# Definir los vectores en R^3
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

# Matriz formada por los vectores
matrix = np.array([v1, v2, v3])

# Calcular el rango de la matriz
rank = np.linalg.matrix_rank(matrix)

# Verificar si los vectores son linealmente independientes
if rank == len(matrix):
    print("Los vectores son linealmente independientes.")
else:
    print("Los vectores son linealmente dependientes.")
```

```python
# Ejercicio 26
import numpy as np

# Definir un conjunto de vectores en R^3
v1 = np.array([1, 2, 3])
v2 = np.array([3, 5, 6])
v3 = np.array([7, 8, 7])

# Matriz formada por los vectores
matrix = np.array([v1, v2, v3])

# Calcular el rango de la matriz
rank = np.linalg.matrix_rank(matrix)

# Verificar si los vectores forman una base en R^3
if rank == 3:
    print("Los vectores forman una base en R^3.")
else:
    print("Los vectores no forman una base en R^3.")
```

```python
# Ejercicio 27
import numpy as np

# Definir los vectores de la base original y la base nueva en R^2
original_base = np.array([[1, 0], [0, 1]])
new_base = np.array([[2, 1], [1, 1]])

# Definir un vector en la base original
vector = np.array([4, 2])

# Calcular las coordenadas del vector en la nueva base
change_of_base_matrix = np.linalg.inv(new_base).dot(original_base)
new_coordinates = change_of_base_matrix.dot(vector)

print("Vector en la base original:", vector)
print("Coordenadas del vector en la nueva base:", new_coordinates)
```

```python
# Ejercicio 28
import numpy as np

# Definir un conjunto de vectores en R^4
v1 = np.array([1, 2, 1])
v2 = np.array([4, 5, 4])
v3 = np.array([7, 7, 9])
v4 = np.array([10, 11, 12])

# Matriz formada por los vectores
matrix = np.array([v1, v2, v3, v4])

# Calcular el rango de la matriz
rank = np.linalg.matrix_rank(matrix)

# La dimensión del espacio generado por los vectores es igual al rango de la matriz
dimension = rank
print("La dimensión del espacio vectorial generado por los vectores es:", dimension)
```

```python
# Ejercicio 30
import numpy as np

# Definir las bases B y B'
B = np.array([[1, 0], [0, 1]])  # Base estándar
B_prime = np.array([[2, 2], [2, 1]])  # Nueva base

# Definir un vector en la base original
v_B = np.array([3, 2])

# Calcular la matriz de cambio de base de B a B'
P_B_to_B_prime = np.linalg.inv(B_prime).dot(B)

# Transformar las coordenadas del vector a la nueva base
v_B_prime = P_B_to_B_prime.dot(v_B)

print("Coordenadas del vector en la nueva base B':", v_B_prime)
```

```python
# Ejercicio 35
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir el punto original
P = np.array([1, 1, 1])

# Definir la matriz de rotación en torno al eje z
theta = np.pi / 4  # 45 grados
Rz = np.array([[np.cos(theta), -np.sin(theta), 0],
               [np.sin(theta), np.cos(theta), 0],
               [0, 0, 1]])

# Aplicar la rotación
P_rotated = Rz.dot(P)

# Plotear el punto original y el punto rotado
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(P[0], P[1], P[2], color='b', label='Punto Original')
ax.scatter(P_rotated[0], P_rotated[1], P_rotated[2], color='r', label='Punto Rotado')

# Configuración del gráfico
ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([0, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title('Rotación de un Punto en 3D')
plt.show()
```
