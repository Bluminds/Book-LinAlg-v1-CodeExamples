```python
# Ejercicio 7
import numpy as np

# MATRIZ IDENTIDAD
def matriz_identidad(n):
    return np.eye(n)

# Ejemplo de uso
I = matriz_identidad(4)
print("Matriz Identidad:")
print(I)

# MATRIZ DIAGONAL
def matriz_diagonal(diagonal):
    return np.diag(diagonal)

# Ejemplo de uso
D = matriz_diagonal([1, 2, 3, 4])
print("\nMatriz Diagonal:")
print(D)

# MATRIZ SIMÉTRICA
def matriz_simetrica(A):
    return (A + A.T) / 2

# Ejemplo de uso
A = np.array([[1, 2, 3], [2, 5, 6], [3, 6, 9]])
S = matriz_simetrica(A)
print("\nMatriz Simétrica:")
print(S)

# MATRIZ ANTISIMÉTRICA
def matriz_antisimetrica(A):
    return (A - A.T) / 2

# Ejemplo de uso
A = np.array([[0, 2, -1], [-2, 0, -4], [1, 4, 0]])
AS = matriz_antisimetrica(A)
print("\nMatriz Antisimétrica:")
print(AS)

# MATRIZ TRIANGULAR SUPERIOR
def matriz_triangular_superior(A):
    return np.triu(A)

# Ejemplo de uso
A = np.array([[1, 2, 3], [3, 5, 6], [4, 8, 7]])
U = matriz_triangular_superior(A)
print("\nMatriz Triangular Superior:")
print(U)

# MATRIZ TRIANGULAR INFERIOR
def matriz_triangular_inferior(A):
    return np.tril(A)

# Ejemplo de uso
L = matriz_triangular_inferior(A)
print("\nMatriz Triangular Inferior:")
print(L)
```

```python
# Ejercicio 8
import numpy as np

def inversa_gauss_jordan(A):
    n = A.shape[0]

    # Crear una matriz aumentada [A | I]
    A_aug = np.hstack((A, np.eye(n)))

    # Aplicar operaciones elementales de fila
    for i in range(n):
        if A_aug[i, i] == 0:
            for j in range(i + 1, n):
                if A_aug[j, i] != 0:
                    A_aug[[i, j]] = A_aug[[j, i]]
                    break
        A_aug[i] = A_aug[i] / A_aug[i, i]
        for j in range(n):
            if i != j:
                A_aug[j] = A_aug[j] - A_aug[j, i] * A_aug[i]

    # La parte derecha de la matriz aumentada es la inversa
    A_inv = A_aug[:, n:]
    return A_inv

# Ejemplo de uso
A = np.array([[2, 7], [6, 3]])
A_inv = inversa_gauss_jordan(A)

print("Inversa usando el método de Gauss-Jordan:")
print(A_inv)
```

```python
# Ejercicio 9
def inversa_adjunta(A):
    det_A = np.linalg.det(A)
    if det_A == 0:
        return None
    cof_A = np.linalg.inv(A).T * det_A  # Obtener cofactores
    adj_A = cof_A.T  # Transponer los cofactores para obtener la adjunta
    A_inv = adj_A / det_A
    return A_inv

# Ejemplo de uso
A = np.array([[2, 7], [6, 3]])
A_inv = inversa_adjunta(A)

print("\nInversa usando el método de adjunto y determinantes:")
print(A_inv)
```

```python
# Ejercicio 10
from scipy.linalg import lu
import numpy as np

def inversa_LU(A):
    P, L, U = lu(A)
    U_inv = np.linalg.inv(U)
    L_inv = np.linalg.inv(L)
    P_inv = np.linalg.inv(P)
    A_inv = np.dot(U_inv, np.dot(L_inv, P_inv))
    return A_inv

# Ejemplo de uso
A = np.array([[2, 7], [6, 3]])
A_inv = inversa_LU(A)

print("\nInversa usando la factorización LU:")
print(A_inv)
```

```python
# Ejercicio 11
import numpy as np
import sympy

def rango_escalonada(A):
    A_sympy = sympy.Matrix(A)
    rref, pivots = A_sympy.rref()
    return len(pivots)

# Ejemplo de uso
A = np.array([[1, 2, 7], [4, 6, 6], [1, 8, 9]])
rango_A = rango_escalonada(A)

print("\nRango usando reducción a forma escalonada fila:")
print(rango_A)
```

```python
# Ejercicio 12
def teorema_rouche_capelli(A, B):
    AB = np.hstack((A, B))
    rango_A = np.linalg.matrix_rank(A)
    rango_AB = np.linalg.matrix_rank(AB)
    if rango_A == rango_AB:
        if rango_A == A.shape[1]:
            return "Sistema compatible determinado (una única solución)"
        else:
            return "Sistema compatible indeterminado (infinitas soluciones)"
    else:
        return "Sistema incompatible (sin solución)"

# Ejemplo de uso
A = np.array([[2, 1, 1], [3, 1, -4], [1, 2, -1]])
B = np.array([[2], [5], [0]])
resultado = teorema_rouche_capelli(A, B)

print("\nResultado del Teorema de Rouché-Capelli:")
print(resultado)
```

```python
# Ejercicio 13
import numpy as np

# Definición de la matriz A
A = np.array([[1, 7, 8],
              [1, 5, 10]])

# Cálculo de la transpuesta de A
A_T = A.T
print("Matriz A:")
print(A)
print("\nTranspuesta de A:")
print(A_T)

# Propiedades de la matriz transpuesta
# Doble transposición
A_T_T = A_T.T
print("\nDoble transposición (debería ser igual a A):")
print(A_T_T)

# Transpuesta de una suma
B = np.array([[4, 3, 1],
              [6, 2, 7]])
sum_transpose = (A + B).T
transpose_sum = A.T + B.T
print("\nTranspuesta de la suma (A + B)^T:")
print(sum_transpose)
print("Suma de las transpuestas A^T + B^T:")
print(transpose_sum)

# Transpuesta de un producto escalar
c = 3
scalar_transpose = (c * A).T
transpose_scalar = c * A.T
print("\nTranspuesta del producto escalar (cA)^T:")
print(scalar_transpose)
print("Producto del escalar y la transpuesta cA^T:")
print(transpose_scalar)

# Transpuesta de un producto de matrices
C = np.array([[1, 2],
              [7, 4],
              [5, 7]])
product_transpose = (A.dot(C)).T
transpose_product = C.T.dot(A.T)
print("\nTranspuesta del producto de matrices (AC)^T:")
print(product_transpose)
print("Producto de las transpuestas C^T A^T:")
print(transpose_product)

# Ejemplo de matriz simétrica
D = np.array([[1, 2, 3],
              [2, 5, 1],
              [3, 1, 7]])
D_T = D.T
is_symmetric = np.array_equal(D, D_T)
print("\nMatriz D:")
print(D)
print("Transpuesta de D:")
print(D_T)
print(f"Es D una matriz simétrica? {is_symmetric}")

# Propiedades de matrices simétricas
# Diagonalización ortogonal
eigvals, eigvecs = np.linalg.eigh(D)
D_diag = eigvecs @ np.diag(eigvals) @ eigvecs.T
print("\nAutovalores de D:")
print(eigvals)
print("Autovectores de D:")
print(eigvecs)
print("Diagonalización ortogonal de D (debería ser igual a D):")
print(D_diag)
```

```python
# Ejercicio 14
import numpy as np
from scipy.linalg import lu, qr, svd, cholesky, eig, schur, hessenberg, polar

# Matriz de ejemplo
A = np.array([[7, 3], [5, 3]])

# DESCOMPOSICIÓN LU
P, L, U = lu(A)
print("Descomposición LU:")
print("P:\n", P)
print("L:\n", L)
print("U:\n", U)

# DESCOMPOSICIÓN QR
Q, R = qr(A)
print("\nDescomposición QR:")
print("Q:\n", Q)
print("R:\n", R)

# DESCOMPOSICIÓN EN VALORES SINGULARES (SVD)
U, S, Vt = svd(A)
print("\nDescomposición SVD:")
print("U:\n", U)
print("S:\n", S)
print("Vt:\n", Vt)

# DESCOMPOSICIÓN DE CHOLESKY (Sólo para matrices positivas definidas)
A_positive = np.dot(A.T, A)
L = cholesky(A_positive)
print("\nDescomposición de Cholesky:")
print("L:\n", L)

# DESCOMPOSICIÓN EIGEN (Autovalores y Autovectores)
vals, vecs = eig(A)
print("\nDescomposición Eigen:")
print("Autovalores:\n", vals)
print("Autovectores:\n", vecs)

# DESCOMPOSICIÓN SCHUR
Q, T = schur(A)
print("\nDescomposición Schur:")
print("Q:\n", Q)
print("T:\n", T)

# DESCOMPOSICIÓN DE HESSENBERG
H, Q = hessenberg(A, calc_q=True)
print("\nDescomposición de Hessenberg:")
print("H:\n", H)
print("Q:\n", Q)

# DESCOMPOSICIÓN DE POLAR
U, P = polar(A)
print("\nDescomposición Polar:")
print("U:\n", U)
print("P:\n", P)

# DESCOMPOSICIÓN BIDIAGONAL (SVD basada)
B, lower, upper = svd(A, full_matrices=False)
print("\nDescomposición Bidiagonal:")
print("B:\n", B)
print("Lower bidiagonal:\n", lower)
print("Upper bidiagonal:\n", upper)
```
