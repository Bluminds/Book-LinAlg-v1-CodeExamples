```python
# Ejercicio 5
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones correspondientes a las ecuaciones
def eq1(x):
    return (4 - 2*x) / 3

def eq2(x):
    return 4*x - 12

x_solution = 20 / 7
y_solution = 4 * (20 / 7) - 12

# Imprimimos las soluciones
print("La solución del sistema es: x =", x_solution, ", y =", y_solution)

# Generación de datos para las gráficas
x_values = np.linspace(-5, 5, 400)
y_values_eq1 = eq1(x_values)
y_values_eq2 = eq2(x_values)

# Gráfica de las ecuaciones
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_eq1, label=r'$2x + 3y = 4$')
plt.plot(x_values, y_values_eq2, label=r'$4x - y = 12$')

# Marcamos la solución en la gráfica
plt.plot(x_solution, y_solution, 'ro')  # punto de solución
plt.text(x_solution, y_solution, f'({x_solution:.2f}, {y_solution:.2f})', fontsize=12, verticalalignment='bottom')

# Configuramos la gráfica
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Resolución del Sistema de Ecuaciones Lineales por el Método de Sustitución')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)
plt.xlim(-5, 5)
plt.show()
```

```python
# Ejercicio 6
import numpy as np
import matplotlib.pyplot as plt

# Definimos la matriz aumentada del sistema de ecuaciones
A = np.array([[1, 2, 1, 9],
              [2, 3, 3, 21],
              [1, 1, 4, 8]], dtype=float)

# Función para realizar la eliminación de Gauss-Jordan
def gauss_jordan(A):
    rows, cols = A.shape
    for i in range(rows):
        # Hacer que el elemento diagonal sea 1
        A[i] = A[i] / A[i, i]
        for j in range(rows):
            if i != j:
                A[j] = A[j] - A[j, i] * A[i]
    return A

# Aplicamos el método de Gauss-Jordan
A_reduced = gauss_jordan(A)

# Leer las soluciones
x = A_reduced[0, -1]
y = A_reduced[1, -1]
z = A_reduced[2, -1]
print("La matriz reducida es:")
print(A_reduced)
print("\nLas soluciones del sistema son:")
print(f"x = {x}, y = {y}, z = {z}")

# Definir valores de z fijos para cada ecuación
z1, z2, z3 = 1, 1, 1

# Funciones de y en términos de x con z fijos
y_values_eq1 = (9 - x_values - z1) / 2
y_values_eq2 = (21 - 2*x_values - 3*z2) / 3
y_values_eq3 = (8 - x_values - 4*z3) / 1

# Gráfica de las ecuaciones
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_eq1, label=r'$x + 2y + z = 9$')
plt.plot(x_values, y_values_eq2, label=r'$2x + 3y + 3z = 21$')
plt.plot(x_values, y_values_eq3, label=r'$x + y + 4z = 8$')

# Configuramos la gráfica
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('Sistema de Ecuaciones Lineales')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.show()
```

```python
# Ejercicio 14
import numpy as np
import scipy.linalg

# Definir la matriz A y el vector b
A = np.array([[2, 7],
              [9, 1]])
b = np.array([5, 10])

# Realizar la descomposición LU
P, L, U = scipy.linalg.lu(A)
print("Matriz L:")
print(L)
print("\nMatriz U:")
print(U)

# Resolver Ly = b
y = scipy.linalg.solve(L, b)

# Resolver Ux = y
x = scipy.linalg.solve(U, y)
print("\nLa solución del sistema es:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
```

```python
# Ejercicio 15
import numpy as np
import scipy.linalg

# Definir la matriz A y el vector b
A = np.array([[2, 7],
              [9, 1]])
b = np.array([5, 10])

# Realizar la descomposición QR
Q, R = np.linalg.qr(A)
print("Matriz Q:")
print(Q)
print("\nMatriz R:")
print(R)

# Multiplicar ambos lados por Q^T
b_t = np.dot(Q.T, b)

# Resolver Rx = Q^T b
x = scipy.linalg.solve_triangular(R, b_t)
print("\nLa solución del sistema es:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
```

```python
# Ejercicio 16
import numpy as np

# Definir la matriz A y el vector b
A = np.array([[2, 7],
              [9, 1]])
b = np.array([5, 10])

# Realizar la descomposición SVD
U, s, Vt = np.linalg.svd(A)
S = np.diag(s)
print("Matriz U:")
print(U)
print("\nMatriz Sumatoria:")
print(S)
print("\nMatriz V^T:")
print(Vt)

# Resolver Ax = b usando SVD
# x = V (Sumatoria)^(-1) U^T b
S_inv = np.linalg.inv(S)
x = np.dot(Vt.T, np.dot(S_inv, np.dot(U.T, b)))

print("\nLa solución del sistema es:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
```

```python
# Ejercicio 17
import numpy as np
import scipy.linalg

# Definir la matriz A (3x2) y el vector b (3x1)
A = np.array([[2, 3],
              [4, 2],
              [2, 5]])
b = np.array([5, 6, 5])

# Realizar la descomposición QR
Q, R = np.linalg.qr(A, mode='reduced')
print("Matriz Q:")
print(Q)
print("\nMatriz R:")
print(R)

# Multiplicar ambos lados por Q^T
b_t = np.dot(Q.T, b)

# Resolver Rx = Q^T b (usando sólo la parte necesaria de R y b_t)
x = scipy.linalg.solve_triangular(R, b_t[:R.shape[0]])
print("\nLa solución del sistema sobredeterminado es:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
```

```python
# Ejercicio 18
import numpy as np

def gauss_seidel(A, b, x0=None, tol=1e-10, max_iter=1000):
    n = len(b)
    x = np.copy(x0) if x0 is not None else np.zeros(n)
    for _ in range(max_iter):
        x_old = np.copy(x)
        for i in range(n):
            sigma1 = np.dot(A[i, :i], x[:i])
            sigma2 = np.dot(A[i, i+1:], x_old[i+1:])
            x[i] = (b[i] - sigma1 - sigma2) / A[i, i]
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            break
    return x

# Ejemplo de uso
A = np.array([[4, -1, 0, 1], [-1, 4, -1, 1], [1, -1, 4, -1], [1, 0, -1, 3]], dtype=float)
b = np.array([15, 10, 10, 10], dtype=float)
x0 = np.zeros(len(b))
x = gauss_seidel(A, b, x0)

print("Solución Gauss-Seidel:", x)
```

```python
# Ejercicio 19
import numpy as np

def jacobi(A, b, x0=None, tol=1e-10, max_iter=1000):
    n = len(b)
    x = np.copy(x0) if x0 is not None else np.zeros(n)
    for _ in range(max_iter):
        x_old = np.copy(x)
        for i in range(n):
            sigma = np.dot(A[i, :], x_old) - A[i, i] * x_old[i]
            x[i] = (b[i] - sigma) / A[i, i]
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            break
    return x

# Ejemplo de uso
A = np.array([[4, -1, 0, 1], [-1, 4, -1, 1], [1, -1, 4, -1], [1, 0, -1, 3]], dtype=float)
b = np.array([15, 10, 10, 10], dtype=float)
x0 = np.zeros(len(b))
x = jacobi(A, b, x0)

print("Solución Jacobi:", x)
```
